```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset
df = pd.read_csv('/Users/pranatimareddy/Downloads/stocksdataset-1.csv')

# Display basic statistics of the dataset
print(df.describe())

# Visualize correlations between numerical features
correlation_matrix = df[['Open', 'Close', 'High', 'Low']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()

# Visualize distribution of numerical features
plt.figure(figsize=(15, 5))

plt.subplot(1, 4, 1)
sns.histplot(df['Open'], kde=True)
plt.title('Distribution of Open Prices')

plt.subplot(1, 4, 2)
sns.histplot(df['Close'], kde=True)
plt.title('Distribution of Close Prices')

plt.subplot(1, 4, 3)
sns.histplot(df['High'], kde=True)
plt.title('Distribution of High Prices')

plt.subplot(1, 4, 4)
sns.histplot(df['Low'], kde=True)
plt.title('Distribution of Low Prices')

plt.show()

# Visualize distribution of the 'Label' column
plt.figure(figsize=(6, 4))
sns.countplot(x='Label', data=df)
plt.title('Distribution of Labels')
plt.show()


```
